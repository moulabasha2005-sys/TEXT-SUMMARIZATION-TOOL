```python
# ==============================
# TEXT SUMMARIZATION TOOL
# ==============================

# Install required library
!pip install -q spacy
!python -m spacy download en_core_web_sm > /dev/null

import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation

# Load NLP model
nlp = spacy.load("en_core_web_sm")

def summarize_text(text, summary_ratio=0.3):
    doc = nlp(text)

    # Create word frequency dictionary
    word_frequencies = {}
    for word in doc:
        if word.text.lower() not in STOP_WORDS and word.text.lower() not in punctuation:
            word_frequencies[word.text.lower()] = word_frequencies.get(word.text.lower(), 0) + 1

    if not word_frequencies:
        return "Not enough content to summarize."

    # Normalize frequencies
    max_frequency = max(word_frequencies.values())
    for word in word_frequencies:
        word_frequencies[word] = word_frequencies[word] / max_frequency

    # Score sentences
    sentence_scores = {}
    for sent in doc.sents:
        for word in sent:
            if word.text.lower() in word_frequencies:
                sentence_scores[sent] = sentence_scores.get(sent, 0) + word_frequencies[word.text.lower()]

    # Select top sentences
    select_length = max(1, int(len(list(doc.sents)) * summary_ratio))
    sorted_sentences = sorted(sentence_scores, key=sentence_scores.get, reverse=True)
    summary_sentences = sorted(sorted_sentences[:select_length], key=lambda s: s.start)

    summary = " ".join([sent.text for sent in summary_sentences])
    return summary


# ==============================
# USER INPUT SECTION
# ==============================

print("===== TEXT SUMMARIZATION TOOL =====\n")
text = input("Enter/Paste your article text below:\n\n")

if len(text.strip()) == 0:
    print("\nNo text provided.")
else:
    summary = summarize_text(text, summary_ratio=0.3)

    print("\n\n----- ORIGINAL TEXT -----\n")
    print(text)

    print("\n\n----- GENERATED SUMMARY -----\n")
    print(summary)
```
